# 2. Exploratory Data Analysis

```python
C1=T[['platform','metascore', 'rating']]
sns.boxplot(data=C1, x= 'platform', y= 'metascore', showfliers=False)
```

El mayor rango está en el Nintendo 3DS, seguido por el WiiU. El WiiU tiene la media más alta. En menor rango parece estar entre el XONE el PC y el Switch.

La mayor cantidad de outliers existe en el PC.

```python
C1=T[['platform','metascore', 'rating']]
sns.boxplot(data=C1, x= 'platform', y= 'metascore', hue='rating', showfliers=False)
```

```python
sns.histplot(C1['metascore'], bins=10)
```

```python
x= T['platform'].value_counts().index
y= T['platform'].value_counts().values
sns.barplot(x=x, y=y)
plt.xticks(rotation=45)
plt.xlabel('Consola/Plataforma')
plt.ylabel('Frecuencia')
plt.title('Distribución por Consola')
```
